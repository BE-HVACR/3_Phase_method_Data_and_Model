```python
import pandas as pd
import datetime,time

import matplotlib.pyplot as plt
```

```python
def date_parser(x):
    re = time.strptime(x, '%m/%d/%Y %H:%M:%S')
    re = time.strftime('%m/%d/%Y %H:%M:%S',re)
    
    return re
```

```python
dat = pd.read_csv('Vertical_Illum_1121.csv',date_parser = date_parser,parse_dates=[1],index_col = 1)
```

```python
plt.figure(figsize=[8.5,4])
ax = plt.subplot()
plt.plot(dat["2022-11-16 00:00:00":"2022-11-17 00:00:00"].iloc[:,2])
ax.set_title('11-16')

plt.figure(figsize=[8.5,4])
ax = plt.subplot()
plt.plot(dat["2022-11-17 00:00:00":"2022-11-18 00:00:00"].iloc[:,2])
ax.set_title('11-17')

plt.figure(figsize=[8.5,4])
ax = plt.subplot()
plt.plot(dat["2022-11-18 00:00:00":"2022-11-19 00:00:00"].iloc[:,2])
ax.set_title('11-18')


plt.figure(figsize=[8.5,4])
ax = plt.subplot()
plt.plot(dat["2022-11-19 00:00:00":"2022-11-20 00:00:00"].iloc[:,2])
ax.set_title('11-19')
```
